```python

# Drop 'customerID' since it's not useful for count plots
columns_to_plot = ['gender', 'SeniorCitizen', 'Partner', 'Dependents',
                   'tenure', 'PhoneService', 'MultipleLines', 'InternetService',
                   'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                   'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract',
                   'PaperlessBilling', 'PaymentMethod', 'MonthlyCharges',
                   'TotalCharges', 'Churn']

# Set the subplot grid size
cols = 4
rows = len(columns_to_plot) // cols + int(len(columns_to_plot) % cols != 0)

# Plot
plt.figure(figsize=(24, rows * 4))
sns.set(style="whitegrid")

for i, col in enumerate(columns_to_plot, 1):
    plt.subplot(rows, cols, i)
    if df[col].nunique() < 25:  # Count plot for categorical / low unique values
        sns.countplot(data=df, x=col, hue=col, palette="Set2", legend=False,
                      order=df[col].value_counts().index)
        plt.xticks(rotation=45)
    else:  # Histogram for continuous / high unique values
        sns.histplot(data=df, x=col, bins=30, kde=True, color='skyblue')
    plt.title(col)

plt.tight_layout()
plt.show()

```

```python
# Add a new column that shows total payment till now (MonthlyCharges * tenure).
df['TotalPayment'] = df.apply(lambda row: row['MonthlyCharges'] * row['tenure'], axis=1)
df[['MonthlyCharges', 'tenure', 'TotalPayment']].head()
```

```python
# Create two separate dataframes for Male and Female customers and then concatenate them back.
male_df = df[df['gender'] == 'Male']
female_df = df[df['gender'] == 'Female']

combined_df = pd.concat([male_df, female_df], axis=0)
combined_df.shape
```

```python
# Create a small sample dataframe with just 2 columns and merge it with the main dataframe on customerID. (merge)
sample = df[['customerID', 'Churn']].head(10)
other = df[['customerID', 'MonthlyCharges']].head(10)

merged = pd.merge(sample, other, on='customerID')
merged

```

```python
# Categorize customers into 'Low', 'Medium', 'High' spenders based on MonthlyCharges using apply.
def categorize_spending(charge):
    if charge < 35:
        return 'Low'
    elif charge < 70:
        return 'Medium'
    else:
        return 'High'

df['SpenderCategory'] = df['MonthlyCharges'].apply(categorize_spending)
df[['MonthlyCharges', 'SpenderCategory']].head()
```

```python
# How many customers churned vs not churned in each PaymentMethod?
df.groupby('PaymentMethod')['Churn'].value_counts()
```

```python
# Calculate total revenue generated from each InternetService type.
df['Revenue'] = df['MonthlyCharges'] * df['tenure']
df.groupby('InternetService')['Revenue'].sum()
```

```python
 # Categorize customers based on tenure into ‘New’, ‘Loyal’, and ‘Old’.
def label_customer(tenure):
    if tenure < 12:
        return 'New'
    elif tenure < 36:
        return 'Loyal'
    else:
        return 'Old'

df['CustomerCategory'] = df['tenure'].apply(label_customer)
df[['tenure', 'CustomerCategory']].head()
```
